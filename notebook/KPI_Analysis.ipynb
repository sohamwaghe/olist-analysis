```python
 import pandas as pd

df = pd.read_csv("orders_full.csv", parse_dates=["order_purchase_timestamp"])
df.head()
```

```python
#Late delivers
late_orders = df[df['delivery_delay'] > 0]
late_percentage = (len(late_orders) / len(df)) * 100
print(f"Late deliveries: {late_percentage:.2f}%")
```

```python
#total revenue
total_revenue = df['total_order_value'].sum()
print(f"Total revenue: ₹{total_revenue:,.0f}")
```

```python
#top 5 states by revnue
top_states = df.groupby('customer_state')['total_order_value'].sum().sort_values(ascending=False).head(5)
print(top_states)
```

```python
#avg review score
avg_review = df.groupby('product_category_name')['review_score'].mean().sort_values(ascending=False).head(10)
print(avg_review)
```

```python
#most sold product categories
top_categories = df['product_category_name'].value_counts().head(10)
print(top_categories)
```
